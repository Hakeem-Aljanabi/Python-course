```python
from sklearn.model_selection import train_test_split
```

```python
df
```

```python
df = pd.get_dummies(df)
```

```python
df
```

```python
X=df.drop('CO2_Emissions',axis=1)
y=df['CO2_Emissions']
```

```python
X_train, X_test, y_train, y_test = train_test_split(
...     X, y, test_size=0.30, random_state=101)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
X_train
```

```python
to_scale = ['Engine_Size','Cylinders','Fuel_Consumption_in_City(L/100 km)','Fuel_Consumption_in_City_Hwy(L/100 km)','Fuel_Consumption_comb(L/100km)','Smog_Level']
```

```python
X_train[to_scale] = scaler.fit_transform(X_train[to_scale])
```

```python
X_test[to_scale] = scaler.transform(X_test[to_scale])
```

```python
from sklearn.linear_model import LinearRegression
```

```python
ln= LinearRegression()
```

```python
ln.fit(X_train,y_train)
```

```python
predictions = ln.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error
```

```python
print(mean_absolute_error(y_test,predictions))
```

```python
print(np.sqrt(mean_squared_error(y_test,predictions)))
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rf = RandomForestRegressor()
```

```python
rf.fit(X_train,y_train)
```

```python
predict = rf.predict(X_test)
```

```python
rf.score(X_test,y_test)
```

```python
test_output = pd.DataFrame({"actual": y_test, "test_predictions": predict})
```
