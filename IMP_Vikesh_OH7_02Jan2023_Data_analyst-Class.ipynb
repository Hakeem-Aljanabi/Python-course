<div class="alert alert-block alert-danger">

**The order delivery trend look odd. Most of the order seems to be happening at very early hours. You will notice that the `order_delivery_time` is currently measure in the UTC format. We would need to convert it to a local time zone based on the location before we can do our data analysis. We will use `dt.tz_convert()` within pandas to do this. As an argument, we specify the time zone to which it needs to be converted** 
    
<div/>

**We can change the time zones. The options we will choose are : `US/Eastern`, `US/Pacific`, `US/Central`**

```python
chi=df[df['region']=='chi']
```

```python
sf= df[df['region']=='sf']
```

```python
df[df['region']=='nyc']
```

```python
nyc = df[df['region']=='nyc']
```

```python
nyc['order_delivery_time']=nyc['order_delivery_time'].dt.tz_convert('US/eastern')
```

```python
nyc
```

```python
chi['order_delivery_time']=chi['order_delivery_time'].dt.tz_convert('US/Central')
```

```python
sf['order_delivery_time']=sf['order_delivery_time'].dt.tz_convert('US/Pacific')
```

```python
sf['month']=sf['order_delivery_time'].dt.month
sf['year']=sf['order_delivery_time'].dt.year
sf['hour']=sf['order_delivery_time'].dt.hour
```

```python
sf['day']=sf['order_delivery_time'].dt.dayofweek
```

```python
sf
```

```python
chi['month']=chi['order_delivery_time'].dt.month
chi['year']=chi['order_delivery_time'].dt.year
chi['hour']=chi['order_delivery_time'].dt.hour
chi['day']=chi['order_delivery_time'].dt.dayofweek
```

```python
nyc['month']=nyc['order_delivery_time'].dt.month
nyc['year']=nyc['order_delivery_time'].dt.year
nyc['hour']=nyc['order_delivery_time'].dt.hour
nyc['day']=nyc['order_delivery_time'].dt.dayofweek
```

```python
new_df = pd.concat([chi,nyc,sf])
```

```python
sns.barplot(x='hour',y='customer_order_rating',data=new_df)
```

```python
new_df
```

```python
sns.barplot(x='day',y='customer_order_rating',data=new_df,hue='region')
```
